# **Insurance Premium Predictive Model**

```python
# step 1: import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: read data
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Insurance%20Premium.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
# step 3: define y and X
df.columns
```

```python
y = df['Premium']
X = df[['Age', 'Gender', 'BMI', 'Children', 'Smoker', 'Region']]
```

```python
X
```

```python
# ordinal encoding
X.Smoker.value_counts()
```

```python
X.Smoker.replace({'yes':0,'no':1},inplace=True)
```

```python
X.Smoker.value_counts()
```

```python
X = pd.get_dummies(X)
```

```python
X
```

```python
# split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2529)
```

```python
# select model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
# train
model.fit(X_train,y_train)
```

```python
# predict
y_pred = model.predict(X_test)
```

```python
# evaluate
from sklearn.metrics import mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```
